```python
import pygame
import numpy
```

```python
class Game:
    def __init__(self,):
        pygame.init()
        pygame.font.init()
        self.window = pygame.display.set_mode((390,390))
        pygame.display.set_caption('Tic-Tac-Toe')
        pygame.display.set_icon(pygame.image.load('pixels.png'))
        self.font = pygame.font.SysFont('ZenTokyoZoo-Regular.ttf', 64)
        
        # Initialize board
        self.reset()
        # Game loop
        self.running = True
        # importing and resizing player symbols
        cross = pygame.image.load('cross.png')
        cross = pygame.transform.scale(cross, (75, 75))
        circle = pygame.image.load('black-circle.png')
        circle = pygame.transform.scale(circle, (75, 75))
        
        # 'o' or 'x'
        self.current_turn = 'x'
        
        while self.running:
            pygame.time.delay(10)
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    self.running = False
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_SPACE:
                        print(self.board)
                        self.reset()
                # bind events and visuals
                if event.type == pygame.MOUSEBUTTONUP:
                    mouse_pos = pygame.mouse.get_pos()
                    
                    if self.pos0.collidepoint(mouse_pos) == True and self.positions[0] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (46, 47))
                            self.current_turn = 'o'
                            self.board[0][0] = 1
                        else:
                            self.window.blit((circle), (46, 47))
                            self.current_turn = 'x'
                            self.board[0][0] = 2
                        self.positions[0] = False
                        
                    if self.pos1.collidepoint(mouse_pos) == True and self.positions[1] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (156, 47))
                            self.current_turn = 'o'
                            self.board[0][1] = 1
                        else:
                            self.window.blit((circle), (156, 47))
                            self.current_turn = 'x'
                            self.board[0][1] = 2
                        self.positions[1] = False
                        
                    if self.pos2.collidepoint(mouse_pos) == True and self.positions[2] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (266, 47))
                            self.current_turn = 'o'
                            self.board[0][2] = 1
                        else:
                            self.window.blit((circle), (266, 47))
                            self.current_turn = 'x'
                            self.board[0][2] = 2
                        self.positions[2] = False
                        
                    if self.pos3.collidepoint(mouse_pos) == True and self.positions[3] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (46, 157))
                            self.current_turn = 'o'
                            self.board[1][0] = 1
                        else:
                            self.window.blit((circle), (46, 157))
                            self.current_turn = 'x'
                            self.board[1][0] = 2
                        self.positions[3] = False
                        
                    if self.pos4.collidepoint(mouse_pos) == True and self.positions[4] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (156, 157))
                            self.current_turn = 'o'
                            self.board[1][1] = 1
                        else:
                            self.window.blit((circle), (156, 157))
                            self.current_turn = 'x'
                            self.board[1][1] = 2
                        self.positions[4] = False
                        
                    if self.pos5.collidepoint(mouse_pos) == True and self.positions[5] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (266, 157))
                            self.current_turn = 'o'
                            self.board[1][2] = 1
                        else:
                            self.window.blit((circle), (266, 157))
                            self.current_turn = 'x'
                            self.board[1][2] = 2
                        self.positions[5] = False
                        
                    if self.pos6.collidepoint(mouse_pos) == True and self.positions[6] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (46, 267))
                            self.current_turn = 'o'
                            self.board[2][0] = 1
                        else:
                            self.window.blit((circle), (46, 267))
                            self.current_turn = 'x'
                            self.board[2][0] = 2
                        self.positions[6] = False
                        
                    if self.pos7.collidepoint(mouse_pos) == True and self.positions[7] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (156, 267))
                            self.current_turn = 'o'
                            self.board[2][1] = 1
                        else:
                            self.window.blit((circle), (156, 267))
                            self.current_turn = 'x'
                            self.board[2][1] = 2
                        self.positions[7] = False
                        
                    if self.pos8.collidepoint(mouse_pos) == True and self.positions[8] == True:
                        if self.current_turn == 'x':
                            self.window.blit((cross), (266, 267))
                            self.current_turn = 'o'
                            self.board[2][2] = 1
                        else:
                            self.window.blit((circle), (266, 267))
                            self.current_turn = 'x'
                            self.board[2][2] = 2
                        self.positions[8] = False
            self.check_winner()
            if self.winner != 0:
                if self.winner == 1:
                    string = "Player 1 wins"
                else:
                    string = "Player 2 wins"
                self.positions = numpy.zeros(9, dtype=bool)
                text = self.font.render(string, True, (150,0,240), (0,0,0))
                self.window.blit(text, (55,175))
                    
            pygame.display.update()
        pygame.font.quit()
        pygame.quit()
        
    def reset(self,):
        self.positions = numpy.ones(9, dtype=bool)
        self.window.fill((175,0,0))
        
        # player flags: p1 -1 - x
        #               p2 -2 - o
        self.board = numpy.zeros((3,3))
        self.winner = 0
        self.current_turn = 'x'
        
        self.pos0 = pygame.draw.rect(self.window, (248,248,255), (35, 35, 100, 100))
        self.pos1 = pygame.draw.rect(self.window, (248,248,255), (145, 35, 100, 100))
        self.pos2 = pygame.draw.rect(self.window, (248,248,255), (255, 35, 100, 100))
        
        self.pos3 = pygame.draw.rect(self.window, (248,248,255), (35, 145, 100, 100))
        self.pos4 = pygame.draw.rect(self.window, (248,248,255), (145, 145, 100, 100))
        self.pos5 = pygame.draw.rect(self.window, (248,248,255), (255, 145, 100, 100))
        
        self.pos6 = pygame.draw.rect(self.window, (248,248,255), (35, 255, 100, 100))
        self.pos7 = pygame.draw.rect(self.window, (248,248,255), (145, 255, 100, 100))
        self.pos8 = pygame.draw.rect(self.window, (248,248,255), (255, 255, 100, 100))
    def check_winner(self,):
        if (numpy.all(self.board[0] == self.board[0][0]) and self.board[0][0] != 0) or (numpy.all(self.board[1] == self.board[1][0]) and self.board[1][0] != 0) or (numpy.all(self.board[2] == self.board[2][0]) and self.board[2][0] != 0) or (numpy.all(self.board[:,0] == self.board[0][0]) and self.board[0][0] != 0) or (numpy.all(self.board[:,1] == self.board[0][1]) and self.board[0][1] != 0) or (numpy.all(self.board[:,2] == self.board[0][2]) and self.board[0][2] != 0) or (numpy.all(self.board.diagonal() == self.board[0][0]) and self.board[0][0] != 0) or (self.board[0][2] == self.board[1][1] and self.board[1][1] == self.board[2][0] and self.board[2][0] != 0):
            if self.current_turn == 'o':
                self.winner = 1
            else:
                self.winner = 2
            return True
        return False
```
